# detecting each and every face in a picture


### A Haar Cascade is an object detection method used to locate an object of interest in images. The algorithm is trained on a large number of positive and negative samples, where positive samples are images that contain the object of interest

```python
%pylab inline
import cv2
```

```python
image = imread("C:\\Users\\arunV\\Downloads\\download2.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
title("Original Image")
axis("off")
imshow(image)
show()

title("Grayscale Image")
axis("off")
imshow(gray, cmap= pylab.gray())
show()
```

## insterting harrcasscade algoriyhm

```python
cascadePath = "C:\\Users\\arunV\\Documents\\haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascadePath)
faces = faceCascade.detectMultiScale(image)
```

## Number of faces found

```python
print ("Number of faces found", faces.shape[0])
```

## finding each and every person in a image

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
    roi_color = image[y:y + h, x:x + w]
    print("[INFO] Object found. Saving locally.")
    cv2.imwrite(str(w) + str(h) + '_faces.jpg', roi_color)
```

```python
status = cv2.imwrite('faces_detected.jpg', image)
print("[INFO] Image faces_detected.jpg written to filesystem: ", status)
```
